```python
import pandas as pd
```

```python
happy_path = "happy15.csv"
happy = pd.read_csv(happy_path)
happy
#data taken from 2015
```

```python
sad_path = "sad.csv"
sad_df = pd.read_csv(sad_path)
sad_df = sad_df.rename({'country':'Country', 'suicides_no':'# of Suicides', 'suicides/100k pop':'Suicides per 100k',
                       'gdp_per_capita ($)':'GDP per Capita'}, axis=1)
sad_df
```

```python
sad = sad_df[sad_df.year == 2015]
sad = sad.groupby("Country")
sad = sad[['Country', '# of Suicides', 'Suicides per 100k', 'GDP per Capita']]
sad = sad.sum('# of Suicides')
sad = sad.reset_index()
sad
```

```python
merge_df = happy.merge(sad, how = "inner", on= "Country")
merge_df.head()
```

```python
merge_df = merge_df.sort_values(by=['Happiness Rank'])
merge_df.head()
```
